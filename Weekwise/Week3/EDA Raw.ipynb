`question_text`: According to the postulates of Werner for coordination compounds

`question_teacher_description`: The primary valencies are ionisable and represented by dotted line.

`imported_from`:  { "bool": false , "doc_id" : "" },

`question_option`:[
  
- `answer_id` : 3564895, **`answer_text`**: primary valency is ionizable,
  `answer_order` : 1 

- answer_id : 3564896, answer_text: secondary valency is ionizable,
  answer_order : 2

- answer_id : 3564897, answer_text : primary and secondary valencies are non-ionizable, answer_order : 3

- answer_id :3564898, answer_text : only primary valency is non-ionizable., answer_order : 4
